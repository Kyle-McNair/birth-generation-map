```python
import pandas as pd
import geopandas as gpd
import shapely
import os
import matplotlib
```

```python
city = gpd.read_file('UA_Top_Cities.geojson')
city.head()
```

```python
count = 0
for i in city.columns:
    print (count, i)
    count = count + 1
```

```python
urban = []
for index, row in city.iterrows():
    urban.append("Urban_Area")
city["Properties"] = urban
main = []
feature = {}
for index, row in city.iterrows():
    jsonlist = []
    yr2018 = [{"Generation": "Generation Z","Value":(row['GZ_2018']/row['Total']),"Color":"#003d13"},
              {"Generation": "Millennials","Value":(row['ML_2018']/row['Total']),"Color":"#558c58"},
              {"Generation": "Generation X","Value":(row['GX_2018']/row['Total']),"Color":"#fdcf79"},
              {"Generation": "Baby Boomers","Value":(row['BB_2018']/row['Total']),"Color":"#9463ab"},
              {"Generation": "Silent Generation","Value":(row['SG_2018']/row['Total']),"Color":"#652187"}]
    jsonlist.append(yr2018)
    
    yr2013 = [{"Generation": "Generation Z","Value":(row['GZ_2013']/row['Total_1']),"Color":"#003d13"},
              {"Generation": "Millennials","Value":(row['ML_2013']/row['Total_1']),"Color":"#558c58"},
              {"Generation": "Generation X","Value":(row['GX_2013']/row['Total_1']),"Color":"#fdcf79"},
              {"Generation": "Baby Boomers","Value":(row['BB_2013']/row['Total_1']),"Color":"#9463ab"},
              {"Generation": "Silent Generation","Value":(row['SG_2013']/row['Total_1']),"Color":"#652187"}]
    jsonlist.append(yr2013)
    
    change = [{"Generation": "Generation Z","Value":(row['GZ_ch']),"Color":"#003d13","Totalyr2018": row['Total'],"Totalyr2013":row['Total_1'], 'Gen2013':row['GZ_2013']},
              {"Generation": "Millennials","Value":(row['ML_ch']),"Color":"#558c58","Totalyr2018": row['Total'],"Totalyr2013":row['Total_1'], 'Gen2013':row['ML_2013']},
              {"Generation": "Generation X","Value":(row['GX_ch']),"Color":"#fdcf79","Totalyr2018": row['Total'],"Totalyr2013":row['Total_1'], 'Gen2013':row['GX_2013']},
              {"Generation": "Baby Boomers","Value":(row['BB_ch']),"Color":"#9463ab","Totalyr2018": row['Total'],"Totalyr2013":row['Total_1'], 'Gen2013':row['BB_2013']},
              {"Generation": "Silent Generation","Value":(row['SG_ch']),"Color":"#652187","Totalyr2018": row['Total'],"Totalyr2013":row['Total_1'], 'Gen2013':row['SG_2013']}]
    c = row['Urban_Area']
    jsonlist.append(c)
    jsonlist.append(change)
    this = {"City":c,"yr2018":yr2018,"yr2013":yr2013,"change":change}
    main.append({row['Properties']:this})
feature.update({"Feature":main}) 
    

    
import json
bg_json = json.dumps(feature) #dumps the list into a json

## output: write a json file
with open('cities.json','w') as  json_file:
    json_file.write(bg_json)
    print ('Converted to json!')
    json_file.close()
```
