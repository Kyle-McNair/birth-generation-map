# 2018 MSA Data

```python
shp = gpd.read_file("US_cbsa_2018.shp")
shp = shp.to_crs('epsg:4326')
shp.head()
```

```python
## Encoding Type is "ISO-8859-1"
msa = pd.read_csv("MSA_2014_2018.csv", encoding = "ISO-8859-1")
msa = pd.DataFrame(msa)
msa.head()
```

```python
count = 0
for i in msa.columns:
    print (count, i)
    count = count + 1
```

```python
## iterate each row  
genZ = []
millennial = []
genX = []
babyBoomer = []
silentG = []
for index, row in msa.iterrows(): 
    genZ_total = row[7] + row[8]+ row[9] + row[10]+ row[11]+ row[12]+ row[13] + row[31]+ row[32]+ row[33]+ row[34]+ row[35]+ row[36]+ row[37]
    millennial_total = row[14] + row[15] + row[16] + row[38] + row[39] + row[40]
    genX_total = row[17] + row[18] + row[19] + row[20] + row[41] + row[42] + row[43] + row[44]
    bb_total = row[21] + row[22] + row[23] + row[24] + row[25] + row[26] + row[45] + row[46] + row[47] + row[48] + row[49] + row[50]
    sg_total = row[27] + row[28] + row[29] + row[51] + row[52] + row[53]
    genZ.append(genZ_total)
    millennial.append(millennial_total)
    genX.append(genX_total)
    babyBoomer.append(bb_total)
    silentG.append(sg_total)
msa['GenZ'] = genZ
msa['Millennials'] = millennial
msa['GenX'] = genX
msa['BabyBoomer'] = babyBoomer
msa['SilentG'] = silentG
msa.head()
```

```python
join = msa
columns = ['Male', 'Male: Under 5 years', 'Male: 5 to 9 years', 'Male: 10 to 14 years',
         'Male: 15 to 17 years','Male: 18 and 19 years','Male: 20 years','Male: 21 years','Male: 22 to 24 years',
         'Male: 25 to 29 years','Male: 30 to 34 years','Male: 35 to 39 years','Male: 40 to 44 years','Male: 45 to 49 years',
         'Male: 50 to 54 years','Male: 55 to 59 years','Male: 60 and 61 years','Male: 62 to 64 years','Male: 65 and 66 years',
         'Male: 67 to 69 years','Male: 70 to 74 years','Male: 75 to 79 years','Male: 80 to 84 years','Male: 85 years and over',
         'Female','Female: Under 5 years','Female: 5 to 9 years','Female: 10 to 14 years','Female: 15 to 17 years','Female: 18 and 19 years',
         'Female: 20 years','Female: 21 years','Female: 22 to 24 years','Female: 25 to 29 years','Female: 30 to 34 years','Female: 35 to 39 years',
         'Female: 40 to 44 years','Female: 45 to 49 years','Female: 50 to 54 years','Female: 55 to 59 years','Female: 60 and 61 years','Female: 62 to 64 years',
         'Female: 65 and 66 years','Female: 67 to 69 years','Female: 70 to 74 years','Female: 75 to 79 years','Female: 80 to 84 years','Female: 85 years and over'
]
join.drop(columns, inplace=True, axis=1)
join.head()
```

```python
msa_data = shp.merge(join, on='GISJOIN')
```

```python
msa_data.head()
```

```python
from shapely.geometry import Point, MultiPoint, LineString, Polygon, shape

msa_data = gpd.GeoDataFrame(msa_data)
msa_data['lat']=msa_data.centroid.y
msa_data['long']=msa_data.centroid.x
genZ_density = []
millennial_density = []
genX_density = []
babyBoomer_density = []
silentG_density = []
for index, row in msa_data.iterrows(): 
    genZ_density.append(float(row['GenZ'])/float(row['Area']))
    millennial_density.append(float(row['Millennials']/float(row['Area'])))
    genX_density.append(float(row['GenX']/float(row['Area'])))
    babyBoomer_density.append(float(row['BabyBoomer']/float(row['Area'])))
    silentG_density.append(float(row['SilentG']/float(row['Area'])))
msa_data['GenZ_Density'] = genZ_density
msa_data['Millennial_Density'] = millennial_density
msa_data['GenX_Density'] = genX_density
msa_data['BabyBoomer_Density'] = babyBoomer_density
msa_data['SilentG_Density'] = silentG_density
msa_data.head()
```

```python
msa_list = []

for row in msa_data.itertuples(index=True):
    msa = {} #node dict
    msa['Name'] = getattr(row, "NAMELSAD")
    msa['Total'] = getattr(row, "Total")
    msa['Lat'] = getattr(row, "lat")
    msa['Long'] = getattr(row, "long")
    msa['GenZ'] =  getattr(row, "GenZ")
    msa['Millennials'] = getattr(row, "Millennials")
    msa['GenX'] = getattr(row, "GenX")
    msa['BabyBoomers'] = getattr(row, "BabyBoomer")
    msa['silentG'] = getattr(row, 'SilentG')
    msa['GenZ_Density'] = getattr(row, 'GenZ_Density')
    msa['Millennials_Density'] = getattr(row, 'Millennial_Density')
    msa['GenX_Density'] = getattr(row, 'GenX_Density')
    msa['BabyBoomers_Density'] = getattr(row, 'BabyBoomer_Density')
    msa['silentG_Density'] = getattr(row, 'SilentG_Density')
    msa_list.append(msa)

import json
msa_json = json.dumps(msa_list) #dumps the list into a json

## output: write a json file
with open('msa_2018.json','w') as  json_file:
    json_file.write(msa_json)
    print ('list is converted to json!')
    json_file.close()
```

```python
msa_json_load = json.loads(msa_json) #load the json
geojson = {"type":"FeatureCollection"} 
features = []
for i in msa_json_load: 
    properties = {} #properties dictionary
    centroids = []
    name = (i['Name'])
    long = (i["Long"]) #finds the latitude in the json
    lat = (i["Lat"]) #finds the longitude in the json
    centroids.append(long)
    centroids.append(lat)
    total = (i['Total'])
    genZ = (i['GenZ'])
    millennial = (i['Millennials'])
    genX = (i['GenX'])
    babyBoomer = (i['BabyBoomers'])
    silentG = (i['silentG'])
    genZ_d = (i['GenZ_Density'])
    millennial_d = (i['Millennials_Density'])
    genX_d = (i['GenX_Density'])
    babyBoomer_d = (i['BabyBoomers_Density'])
    silentG_d = (i['silentG_Density'])
    msa_dict = {"type":"Feature","geometry":{"type":"Point","coordinates":centroids},"properties":{"MSA": name, "Total": total, "GenZ":genZ, "GenZ_Density": genZ_d,
                                                                                                       "Millennials": millennial, "Millennials_Density": millennial_d,
                                                                                                       "GenX": genX, "GenX_Density": genX_d, "BabyBoomer":babyBoomer, 
                                                                                                       "BabyBoomers_Density": babyBoomer_d, "SilentG":silentG,
                                                                                                      "SilentG_Density": silentG_d}}
    features.append(msa_dict)

    
geojson.update({"features":features})

msa_geojson = json.dumps(geojson) #dump the json into a geojson

with open('msa_2018_geojson.json','w') as  geojson_file:
    geojson_file.write(msa_geojson)
    print ('json is converted to geojson!')
```
