```python
## Encoding Type is "ISO-8859-1"
bg = pd.read_csv("2018_ACS_BG.csv", encoding = "ISO-8859-1")
bg = pd.DataFrame(bg)
bg.head()
```

```python
genZ = []
millennial = []
genX = []
babyBoomer = []
silentG = []
for index, row in bg.iterrows(): 
    genZ_total = row[5] + row[6]+ row[7] + row[8]+ row[9]+ row[10]+ row[11] + row[29]+ row[30]+ row[31]+ row[32]+ row[33]+ row[34]+ row[35]
    millennial_total = row[12] + row[13] + row[14] + row[36] + row[37] + row[38]
    genX_total = row[15] + row[16] + row[17] + row[18] + row[39] + row[40] + row[41] + row[42]
    bb_total = row[19] + row[20] + row[21] + row[22] + row[23] + row[24] + row[43] + row[44] + row[45] + row[46] + row[47] + row[48]
    sg_total = row[25] + row[26] + row[27] + row[49] + row[50] + row[51]
    genZ.append(genZ_total)
    millennial.append(millennial_total)
    genX.append(genX_total)
    babyBoomer.append(bb_total)
    silentG.append(sg_total)
    if index == (0.25*(len(x))):
        print("25% Complete")
    if index == (0.50*(len(x))):
        print("50% Complete")
    if index == (0.75*(len(x))):
        print("75% Complete")
bg['GZ'] = genZ
bg['ML'] = millennial
bg['GX'] = genX
bg['BB'] = babyBoomer
bg['SG'] = silentG
bg.head()
```

```python
Total = bg['Total'].sum()
GZ_Total = bg['GZ'].sum()
ML_Total = bg['ML'].sum()
GX_Total = bg['GX'].sum()
BB_Total = bg['BB'].sum()
SG_Total = bg['SG'].sum()
```

```python
print(Total, GZ_Total, ML_Total, GX_Total, BB_Total, SG_Total)
```

```python
GZ_sh = GZ_Total/Total
ML_sh = ML_Total/Total
GX_sh = GX_Total/Total
BB_sh = BB_Total/Total
SG_sh = SG_Total/Total
print(GZ_sh, ML_sh, GX_sh, BB_sh, SG_sh)
```

```python
bg_sh = {"GZ":GZ_sh,"ML":ML_sh,"GX":GX_sh,"BB":BB_sh,"SG":SG_sh}

import json
bg_json = json.dumps(bg_sh) #dumps the list into a json

## output: write a json file
with open('bg_share.json','w') as  json_file:
    json_file.write(bg_json)
    print ('Converted to json!')
    json_file.close()
```
